```python
# 3) Barras: frecuencias de Sí/No para Python/Colab/GitHub
# ====================================================
fig, axes = plt.subplots(1,3, figsize=(12,4), sharey=True)
for ax, col, title, color in zip(
    axes,
    ['prog_python','usa_colab','cuenta_github'],
    ['Programó en Python','Usó Google Colab','Cuenta de GitHub'],
    [accent, accent2, accent3]
):
    counts = df[col].value_counts().reindex([True, False]).fillna(0).astype(int)
    labels = ['Sí','No']
    ax.bar(labels, counts, color=[color, neutral], edgecolor='white')
    ax.set_title(title)
    ax.set_ylim(0, max(counts.max(), 1)+2)
    for i, v in enumerate(counts):
        ax.text(i, v+0.1, str(v), ha='center', va='bottom', fontsize=9)
axes[0].set_ylabel('Número de estudiantes')
plt.tight_layout()
plt.show()
fig.savefig('03_barras_frecuencias.png', dpi=180)
```

```python
# 4) Mapas de calor 2x2 para combinaciones
# =======================================
pairs = [
    ('prog_python','usa_colab','Python vs Colab'),
    ('prog_python','cuenta_github','Python vs GitHub'),
    ('usa_colab','cuenta_github','Colab vs GitHub')
]

for idx, (a,b,titulo) in enumerate(pairs, start=1):
    a_bool = df[a].to_numpy(dtype=bool)
    b_bool = df[b].to_numpy(dtype=bool)
    t = np.zeros((2,2), dtype=int)
    t[1,1] = np.count_nonzero(a_bool & b_bool)
    t[1,0] = np.count_nonzero(a_bool & ~b_bool)
    t[0,1] = np.count_nonzero(~a_bool & b_bool)
    t[0,0] = np.count_nonzero(~a_bool & ~b_bool)

    fig, ax = plt.subplots(figsize=(5,4))
    im = ax.imshow(t, cmap='Blues')
    ax.set_title(f'Tabla 2×2: {titulo}')
    ax.set_xticks([0,1]); ax.set_xticklabels(['No','Sí'])
    ax.set_yticks([0,1]); ax.set_yticklabels(['No','Sí'])
    ax.set_xlabel(b); ax.set_ylabel(a)

    for i in range(2):
        for j in range(2):
            ax.text(j, i, t[i,j], ha='center', va='center',
                    color='black', fontsize=12, fontweight='bold')

    cbar = fig.colorbar(im, ax=ax)
    cbar.ax.set_ylabel('Conteo', rotation=270, labelpad=12)
    plt.tight_layout()
    plt.show()
    fig.savefig(f'04_heatmap_{idx}.png', dpi=180)
```

```python
# 5) Barras: edad promedio por grupo (Sí/No) por variable
# ======================================================
fig, axes = plt.subplots(1,3, figsize=(12,4), sharey=True)
for ax, col, title, color in zip(
    axes,
    ['prog_python','usa_colab','cuenta_github'],
    ['Edad promedio por Python','Edad promedio por Colab','Edad promedio por GitHub'],
    [accent, accent2, accent3]
):
    means = df.groupby(col)['edad'].mean().reindex([True, False])
    labels = ['Sí','No']
    bars = ax.bar(labels, means, color=[color, neutral], edgecolor='white')
    for i, v in enumerate(means):
        ax.text(i, v+0.05, f"{v:.2f}", ha='center', va='bottom', fontsize=9)
    ax.set_title(title)
    ax.set_xlabel('Grupo')
axes[0].set_ylabel('Edad promedio (años)')
plt.tight_layout()
plt.show()
fig.savefig('05_barras_edad_promedio.png', dpi=180)
```

MSE y otros indicadores de desempeño de modelos.

```python
import numpy as np
import pandas as pd
```

```python
anios_escolaridad = np.array([0, 3, 6, 9, 12, 16, 18])
ingreso_mensual_estimado_mx = np.array([12621, 15723, 19587, 24401, 30398, 40747, 47176])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx)
plt.title('Años de Escolaridad vs. Ingreso Mensual Estimado')
plt.xlabel('Años de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.show()
```

Primer intento de modelar los años de escolaridad y el ingreso estimado.   
Utilizar el primer punto y el útlimo para trazar una recta:   $$ m = \frac{y_{2}-y_{1}} {x_{2}-x_{1}} $$
$$ b = y_{1} - m*x_{1} $$

```python
x1 = anios_escolaridad[0]
y1 = ingreso_mensual_estimado_mx[0]
x2 = anios_escolaridad[-1]
y2 = ingreso_mensual_estimado_mx[-1]

m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1

print(f"La ecuación de la recta es: y = {m:.2f}x + {b:.2f}")
```

```python
linea_ingreso_estimado_new_plot = m * anios_escolaridad + b
linea_ingreso_estimado_new_plot
```

```python
import matplotlib.pyplot as plt

# Calcular los valores y de la recta utilizando m y b
linea_ingreso_estimado_new_plot = m * anios_escolaridad + b

#Imprime los datos originales
plt.figure(figsize=(10, 6))
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx, label='Datos Originales')
#Genera la recat de regresión en color rojo
plt.plot(anios_escolaridad, linea_ingreso_estimado_new_plot, color='red',
         label=f'Recta de Regresión (y = {m:.2f}x + {b:.2f})')
plt.title('Años de Escolaridad vs. Ingreso Mensual Estimado con Línea de Regresión')
plt.xlabel('Años de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.legend()
plt.show()
```

Valores predichos
: Son los ingresos mensuales que nuestra recta de regresión predice para cada año de escolaridad.

Errores (residuales): Es la diferencia entre el ingreso mensual real (ingreso_mensual_estimado_mx) y el ingreso mensual predicho (y_pred). Un error positivo significa que la predicción fue menor que el valor real, y un error negativo indica que la predicción fue mayor.

Errores al cuadrado: Hemos elevado al cuadrado cada uno de los errores. Esto se hace para eliminar los signos negativos (de modo que los errores positivos y negativos no se cancelen entre sí) y para penalizar más los errores grandes.

Suma de errores cuadrados: Es la suma de todos los errores al cuadrado. Este valor nos da una medida total de la 'distancia' de nuestros puntos a la línea.

Error Cuadrático Medio (MSE): Es la suma de los errores cuadrados dividida por el número total de puntos de datos. El MSE es una medida promedio de la magnitud de los errores. Un MSE más bajo indica que el modelo se ajusta mejor a los datos.

En este caso, el Error Cuadrático Medio (MSE) es de aproximadamente 13,196,667.31. Este número nos da una idea del rendimiento general de la recta que trazamos, mostrando la magnitud promedio de los errores al cuadrado entre los valores observados y los valores predichos por la línea.

Fórmula del MSE (Error Cuadrático Medio) del Inglés Mean Squared Error

$$MSE = \frac{Σ_{i=1}^{n}(Y_{i}-\hat{Y_{i}})^{2}} {n} $$

$MSE = $ error cuadrátcio medio   
$n = $ Número de datos   
$Y_{i} = $ Valores observados   
$(\hat{Y_{i})} = $ Valores predichos   


```python
# 1. Calcular los valores predichos (y_pred) usando la ecuación de la recta
y_pred = m * anios_escolaridad + b
y_pred
```

```python
# 2. Calcular los errores (residuales): la diferencia entre los valores reales y los predichos
errores = ingreso_mensual_estimado_mx - y_pred
errores
```

```python
# 3. Elevar al cuadrado cada error
errores_cuadrados = errores**2
errores_cuadrados
```

```python

# 4. Sumar los errores cuadrados
suma_errores_cuadrados = np.sum(errores_cuadrados)
suma_errores_cuadrados
```

```python
# 5. Dividir la suma de los errores cuadrados por el número de puntos para obtener el MSE
mse = suma_errores_cuadrados / len(anios_escolaridad)
mse
```

Ahora comparemos el MSE de una regresión lineal.
Por ahora no es importante el detalle de la regresión lineal.
Lo importante es comparar el MSE.

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Cambia la forma de anios_escolaridad a 2D array como es requerido por sklearn
X = anios_escolaridad.reshape(-1, 1)
y = ingreso_mensual_estimado_mx

X
```

```python
# Crear y ajustar el modelo de regresión lineal
modelo_regresion = LinearRegression()
modelo_regresion.fit(X, y)

# Obtener los coeficientes (m) e intercepto (b) del nuevo modelo
m_sklearn = modelo_regresion.coef_[0]
b_sklearn = modelo_regresion.intercept_

print(f"Coefficients (m) from scikit-learn: {m_sklearn:.2f}")
print(f"Intercept (b) from scikit-learn: {b_sklearn:.2f}")
```

El nuevo modelo de recta obtenido con regresión lineal es:   
$$ y =1914.55 * x + 9731.66 $$

```python
#Grafica la línea del modelo de regresión lineal de sklearn junto con los datos originales
# 1. Calculamos las predicciones del modelo de Sklearn
y_pred_sklearn = m_sklearn * anios_escolaridad + b_sklearn

plt.figure(figsize=(10, 6))

# Datos originales
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx, color='blue', label='Datos Originales', zorder=5)
plt.plot(anios_escolaridad, linea_ingreso_estimado_new_plot, color='red', linestyle='--', alpha=0.6, label='Recta Manual (Punto A a B)')
# Recta Sklearn
plt.plot(anios_escolaridad, y_pred_sklearn, color='black', linewidth=3, label='Recta Sklearn (Optimizado)')

plt.title('Comparativa: Selección Manual vs. Regresión Lineal de Sklearn')
plt.xlabel('Años de Escolaridad')
plt.ylabel('Ingreso Mensual (MXN)')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```

```python
#Calcula el MSE de esta regresión lineal y compara con el resultado anterior
errores_sklearn = ingreso_mensual_estimado_mx - y_pred_sklearn
mse_sklearn = np.mean(errores_sklearn**2)

print(f"MSE del primer intento (Manual): {mse:.2f}")
print(f"MSE de Sklearn (Optimizado):     {mse_sklearn:.2f}")
```

 ¿Cuál recta tiene menos error con respecto a los datos originales?
 La regresión lineal de Sklearn utiliza un método llamado Mínimos Cuadrados Ordinarios (OLS). En lugar de forzar el paso por los extremos, busca una posición "promedio" que minimice la distancia al cuadrado de todos los puntos simultáneamente. Aunque la línea de Sklearn no toca exactamente el primer ni el último punto, está mucho más cerca del grupo total de datos.

Indicadores de Error del ModeloValores Predichos ($\hat{y}$): Es la estimación que hace el modelo para cada valor de $x$.$$\hat{y} = mx + b$$Errores o Residuales ($e$): La diferencia entre el valor real y el predicho.$$e = y - \hat{y}$$Errores al Cuadrado ($e^2$): Elevamos el error al cuadrado para penalizar desviaciones grandes y eliminar signos negativos.$$e^2 = (y - \hat{y})^2$$Suma de Errores Cuadrados (SSE): La acumulación total de todas las desviaciones al cuadrado.$$SSE = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$Error Cuadrático Medio (MSE): El promedio de los errores cuadrados. Es nuestra principal métrica de desempeño.$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

```python
#Calcula los indicadores para la región lineal
# 1. Valores predichos por el modelo de sklearn
y_pred_sklearn = modelo_regresion.predict(X)

# 2. Calcular los residuales (errores)
errores_sklearn = ingreso_mensual_estimado_mx - y_pred_sklearn

# 3. Errores al cuadrado
errores_cuadrados_sklearn = errores_sklearn**2

# 4. Suma de errores cuadrados (SSE)
sse_sklearn = np.sum(errores_cuadrados_sklearn)

# 5. Error Cuadrático Medio (MSE)
mse_sklearn = sse_sklearn / len(anios_escolaridad)

# Mostramos los resultados finales
print(f"Indicadores de la Regresión Sklearn:")
print(f"------------------------------------")
print(f"SSE: {sse_sklearn:,.2f}")
print(f"MSE: {mse_sklearn:,.2f}")
```

Razón (media, mediana, IQR, coeficiente de variación):

```python
def resumen_razon(s): #FUNCIÓN
    return pd.Series({ #Regresa una serie de datos
    "n": s.count(),
    "media": s.mean(),
    "mediana": s.median(),
    "sd": s.std(),
    "min": s.min(),
    "q1": s.quantile(0.25),
    "q3": s.quantile(0.75),
    "max": s.max(),
    "cv": s.std()/s.mean() if s.mean()!=0 else np.nan
    }).round(2)

razon_stats = pd.concat({ #Lo concatena en columnas
    "edad": resumen_razon(df["edad"]),
    "ingreso_mensual": resumen_razon(df["ingreso_mensual"]),
    "visitas_ultimo_anio": resumen_razon(df["visitas_ultimo_anio"]),
    "dias_desde_ultima_visita": resumen_razon(df["dias_desde_ultima_visita"])
    }, axis=1)
razon_stats
```

```python
#Track blue color usage over time
features = pd.read_csv('mondrian-painting-features.csv')
features.head()
painting_info = pd.read_csv('mondrian-painting-info.csv')
painting_info.head()

features["feature_area"] = features["width"] * features["height"]
total_area = features.groupby("painting_id")["feature_area"].sum().reset_index()
total_area.rename(columns={"feature_area":"painting_area"}, inplace=True)
blue_features = features[features["color"] == "blue"] #Filtramos
#Realizamos calculos de procentaje para el color azul
blue_area = blue_features.groupby("painting_id")["feature_area"].sum().reset_index()
blue_area.rename(columns={"feature_area":"blue_area"}, inplace=True)
painting_info = painting_info.merge(total_area, on="painting_id")
painting_info = painting_info.merge(blue_area, on="painting_id", how="left")
painting_info["blue_area"].fillna(0)
painting_info["blue_porcent"] = painting_info["blue_area"] / painting_info["painting_area"] * 100
#Graficar azul
plt.scatter(painting_info["year"], painting_info["blue_porcent"], color="blue")
plt.xlabel("Año")
plt.ylabel("% Azul")
plt.title("Porcentaje de azul en pinturas de Mondrian")
plt.show()

# Calculamos los porcentajes de los demas colores
colors = ["red","yellow","white","black","gray"]
for c in colors:
    color_features = features[features["color"] == c]
    color_area = color_features.groupby("painting_id")["feature_area"].sum().reset_index()
    color_area.rename(columns={"feature_area":f"{c}_area"}, inplace=True)
    painting_info = painting_info.merge(color_area, on="painting_id", how="left")
    painting_info[f"{c}_area"].fillna(0)
    painting_info[f"{c}_porcent"] = painting_info[f"{c}_area"] / painting_info["painting_area"] * 100

#Graficamos todos los colores restantes juntos
for c in colors:
    plt.scatter(painting_info["year"], painting_info[f"{c}_porcent"])
plt.xlabel("Año")
plt.ylabel("% del color en la pintura")
plt.title("Evolución del uso de colores en Mondrian")
plt.show()
```

Application: Exploring Hand-written Digits

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.images.shape
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(10, 10, figsize=(8, 8),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))

for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='binary', interpolation='nearest')
    ax.text(0.05, 0.05, str(digits.target[i]),
            transform=ax.transAxes, color='green')
```

```python
X = digits.data
X.shape
```

```python
y = digits.target
y.shape
```

Unsupervised learning: Dimensionality reduction

```python
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
iso.fit(digits.data)
data_projected = iso.transform(digits.data)
data_projected.shape
```

```python
plt.scatter(data_projected[:, 0], data_projected[:, 1], c=digits.target,
            edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('Spectral', 10))
plt.colorbar(label='digit label', ticks=range(10))
plt.clim(-0.5, 9.5);
```

Classification on digits

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(Xtrain, ytrain)
y_model = model.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

```python
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(ytest, y_model)

sns.heatmap(mat, square=True, annot=True, cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value');
```
